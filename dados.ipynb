```python
import requests
import pandas as pd

# URL da API
url = "https://apps.tre-pe.jus.br/locaisVotacao/locais"

# Função para obter locais de votação e criar um DataFrame
def obter_locais_votacao():
    try:
        # Fazendo a requisição GET para o endpoint da API
        response = requests.get(url)
        
        # Verificando se a requisição foi bem-sucedida
        if response.status_code == 200:
            # Convertendo a resposta JSON para um dicionário Python
            data = response.json()
            
            # Criando um DataFrame a partir dos dados
            df = pd.DataFrame(data)
            return df
        else:
            print(f"Erro ao acessar a API: {response.status_code}")
            return None
    except Exception as e:
        print(f"Ocorreu um erro: {e}")
        return None

# Chamar a função e armazenar o DataFrame
locais_votacao_df = obter_locais_votacao()
if locais_votacao_df is not None:
    olinda_df = locais_votacao_df.loc[locais_votacao_df['municipio'] == 'OLINDA']
else:
    print("Não foi possível obter os dados dos locais de votação.")
```

```python
import requests
# URL da API Nominatim
url = 'https://nominatim.openstreetmap.org/search'

# Função para consultar a API Nominatim para um bairro específico
def consultar_bairro(bairro):
    # Parâmetros da pesquisa
    params = {
        'q': f'{bairro} Olinda',
        'format': 'json',
        'addressdetails': 1
    }
    
    # Fazer a solicitação
    response = requests.get(url, params=params)
    
    # Verificar se a solicitação foi bem-sucedida
    if response.status_code == 200:
        data = response.json()
        return data
    else:
        print(f"Erro ao consultar {bairro}: {response.status_code}")
        return None

# Lista para armazenar os resultados
resultados = []

# Loop para consultar todos os bairros únicos e armazenar os resultados
for bairro in olinda_df_unico:
    print(f"Consultando bairro: {bairro}")
    resultado = consultar_bairro(bairro)
    
    if resultado:
        for place in resultado:
            resultados.append({
                'bairro': bairro,
                'nome': place['display_name'],
                'latitude': place['lat'],
                'longitude': place['lon'],
                'detalhes_endereco': place['address']
            })

# Converter a lista de resultados em um DataFrame
resultados_df = pd.DataFrame(resultados)
```
